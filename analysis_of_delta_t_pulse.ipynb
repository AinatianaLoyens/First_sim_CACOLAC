The continuous model performs anyway slightly better with small $T$ because $x$ in the continuous model is decreasing when $E>r$ ($\dot{x} < 0$) while $x$ in the impulsive model can a piecewise increasing function. With a small $T$, when $x_{cont}$ is already below $\epsilon$, $x_{imp}$ is still increasing before the next impulsion. That creates a small delay.
