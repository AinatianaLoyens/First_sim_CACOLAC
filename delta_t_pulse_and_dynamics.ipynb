## 3 Simulations at equilibrium to understand why the optimal $t_{pulse}$ is arround the same value for each $T$

The coexistence equilibrium with $E=0$ is $\left( \frac{m}{\gamma a},\frac{r}{a}\left( 1-\frac{m}{K \gamma a}\right) \right)$

````python
#Equilibrium
x_star_3 = m/(gamma*a)
y_star_3 = r/a * ( 1 - m/(K*gamma*a) )
````

````python
#pcolormesh of the equilibrium
double_exo.plot_t_eta_contour_from_t_pulse_and_T_prop_mortality_on_x(
    xyI0_imp=[x_star_3,y_star_3,0],
    xyI0_cont=[x_star_3,y_star_3,0],
    t=tt,
    gamma=gamma,
    E_c=E_sup,
    func_g=double_exo.logistic_model_x,
    kwargs_g={'r':r, 'K':K},
    func_f=double_exo.multiply_x,
    kwargs_f={'z': a}, 
    func_m=double_exo.identity,
    kwargs_m={'z': m},
    t_0=t_0,
    t_n=200,
    eps=0.01,
    T_start=1,
    T_stop=50,
    T_num=200,
    t_pulse_start=0,
    t_pulse_stop=100,
    t_pulse_num= 201,
    plot_function='pcolormesh',
    plot_T_of_T=True
)
````
