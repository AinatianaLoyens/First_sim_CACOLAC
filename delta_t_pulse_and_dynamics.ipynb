Those results are the opposite of when $(x_0,y_0) = (K,0.1)$.  
Here, with much more predators at initial value, the impulsive model is better when the first pulse is early, and the continuous model is better when the first pulse is late.

It could be because the predators size are now less affected by the pulse. There are still enough to eat the pests even if the pulse is early.  
If we wait, the oscillatory behaviour of the dynamic system makes the pests grow again.
