# Simulations of the initial value of the third periodic solution

When $E < r\left(1-\frac{m}{K \gamma a}\right)$, we don't have the explicit expression of the periodic solution. We excpect that the third periodic solution is stable for $E < r\left(1-\frac{m}{K \gamma a}\right)$ because the first two are unstable. We simulate the initial value of the third periodic solution numerically. It is close to the initial value of the last period of the simulation. We do the the simulation for many values of $E < r\left(1-\frac{m}{K \gamma a}\right)$.  
The first exogenous mortality event is always $t_{pulse} = T$ like in the basic model.

We fix the folowing parameters:  
$r = 0.5$  
$K = 10$  
$a = 0.1$  
$m = 0.1$  
$\gamma = 0.8$  
$T = 5$  
$x_0 = 5$  
$y_0 = 1$

```python
#Import librairies
import numpy as np
import matplotlib.pyplot as plt

#Import the module and make it reload each time there is a change 
import importlib
import double_exo
importlib.reload(double_exo)
```

```python
#Function for the second periodic solution with respect to E
def give_second_x_p_0(E,*,r,K,T):
    '''This function returns the initial value of the second periodic solution with respect to E.
    The goal is to draw the bifurcation diagram with respect to E'''
    x_p_0 = ( np.exp((r-E)*T) - 1 ) * K / ( np.exp(r*T) - 1 )
    return x_p_0
```

```python
#Fixed parameters
r = 0.5
K = 10 
a = 0.1
m = 0.1
gamma = 0.8
T = 0.5

t_0 = 0
t_n = 500

tt = 1 #just to not have a missing argument t when we solve models

#Initial values
x_0 = 5
y_0 = 1
I_0 = 0
xyI_0 = [x_0, y_0, I_0]
```

```python
#Bound: the maximum value of E to search the third periodic solution.
E_bound = r * ( 1 - m/(K*gamma*a) )
```

## 3. Bifurcation diagram

To draw the bifurcation diagram, we first have three intervals of $E$:  
-$E < r\left(1-\frac{m}{K \gamma a}\right)$  
-$r\left(1-\frac{m}{K \gamma a}\right) < E < r$  
-$E > r$

```python
#The three intervals
#The first interval is already low_E
mid_E = np.linspace(E_bound, r, 100)
high_E = np.linspace(r, 3/2*r, 100)
```

First interval: three periodic solutions

```python
x_p_0_first_low = np.zeros_like(low_E) #Initial values of the first periodic solution for low E
x_p_0_second_low = give_second_x_p_0(low_E, r=r, K=K, T=T) #Initial values of the second periodic solution for low E
# The iniial values of the thrid periodic solution is already x_p_0_third_low
```

Second interval: two periodic solutions

```python
x_p_0_first_mid = np.zeros_like(mid_E) #Initial values of the first periodic solution for mid E
x_p_0_second_mid = give_second_x_p_0(mid_E, r=r, K=K, T=T) #Initial values of the second periodic solution for mid E
```

Third interval: one periodic solution

```python
x_p_0_first_high = np.zeros_like(high_E) #Initial values of the first periodic solution for high E
```

Plot of the bifurcation diagram

```python
#First interval (low E)
plt.plot(low_E, x_p_0_first_low, color = 'red') #unstable
plt.plot(low_E, x_p_0_second_low, color = 'red') #unstable
plt.plot(low_E, x_p_0_third_low, color = 'green') #stable
plt.plot(mid_E, x_p_0_first_mid, color = 'red') #unstable
plt.plot(mid_E, x_p_0_second_mid, color = 'green') #stable
plt.plot(high_E, x_p_0_first_high, color = 'green') #stable
plt.xlabel('Effort E')
plt.ylabel('Initial value x_p_0 of the periodic solution')
plt.title('Bifurcation diagram for the impuslive model in two dimensions')
plt.suptitle(f'{r=}, {K=}, {T=}, {m=}, {gamma=}, {a=}')
plt.show()
```
