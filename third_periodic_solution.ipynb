# Simulations of the initial value of the third periodic solution

When $E < r\left(1-\frac{m}{K \gamma a}\right)$, we don't have the explicit expression of the periodic solution. We excpect that the third periodic solution is stable for $E < r\left(1-\frac{m}{K \gamma a}\right)$ before the first two are unstable. We simulate the initial value of the third periodic solution numerically. It is close to the initial value of the last period of the simulation. We do the the simulation for many values of $E < r\left(1-\frac{m}{K \gamma a}\right)$.  
The first exogenous mortality event is always $t_{pulse} = T$ like in the basic model.

We fix the folowing parameters:  
$r = 0.5$  
$K = 10$  
$a = 0.1$  
$m = 0.1$  
$\gamma = 0.8$  
$T = 5$  
$x_0 = 5$  
$y_0 = 1$

```python
#Import librairies
import numpy as np
import matplotlib.pyplot as plt

#Import the module and make it reload each time there is a change 
import importlib
import double_exo
importlib.reload(double_exo)
```

```python
#Fixed parameters
r = 0.5
K = 10 
a = 0.1
m = 0.1
gamma = 0.8
T = 5

t_0 = 0
t_n = 500

tt = 1 #just to not have a missing argument t when we solve models

#Initial values
x_0 = 5
y_0 = 1
I_0 = 0
xyI_0 = [x_0, y_0, I_0]
```

```python
#Bound: the maximum value of E to search the third periodic solution.
E_bound = r * ( 1 - m/(K*gamma*a) )
```
